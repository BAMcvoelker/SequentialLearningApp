```python
from lolopy.learners import RandomForestRegressor
from operator import add
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

def predictions(X_train, y_train, X_test):
    """predict expected target value and the corresponding uncertainty of all X_test points"""
    rf = RandomForestRegressor()
    rf.fit(X_train, y_train)
    y_pred, y_std = rf.predict(X_test, return_std=True)
    return (y_pred, y_std)

# methods for selecting the next best candidate "to make an experiment with"  
def MEI(y_pred):
    """outputs the X_test index with the biggest expected target value corresponding to it"""
    index_new_experiment = y_pred.argmax(axis=0)
    return index_new_experiment 

def MU(y_std):
    """outputs the X_test index with the biggest uncertainty corresponding to it"""
    index_new_experiment = y_std.argmax(axis=0)
    return index_new_experiment

def MLI(y_pred, y_std):
    """outputs the X_test value which gives the maximum likelyhood of improvement"""
    value = list( map(add, y_pred, y_std) )
    value = np.asarray(value)
    index_new_experiment = value.argmax(axis=0)
    return index_new_experiment



def building_new_training_and_test_sets(X_train, y_train, X_test, y_test, index, iteration, x_times_choosen):
    """prepare the next training and testset for the next iteration"""
    #append to the X_train and y_train
    X_t = np.reshape(X_test[index], (1,22))
    X_train = np.append(X_train,X_t, axis= 0)
    y_train = np.append(y_train, y_test[index])
    #drop from X_test and y_test
    X_test = np.delete(X_test, index, 0)
    y_test = np.delete(y_test, index, 0)
    #
    x_times_choosen[index] = x_times_choosen[index] + 1
    iteration = iteration +1
    return (X_train, y_train, X_test, y_test, iteration, x_times_choosen)


def finding_best_candidate(y, y_train, y_test, X_train, X_test, method):
    """finding iteratively the candidate with the optimal target value"""
    iteration = 0
    x_times_choosen = np.zeros(np.shape(y))
    while y_train.max(axis=0) < y.max(axis=0):
        y_pred, y_std = predictions(X_train, y_train, X_test)
        if method == MEI:
            index_new_experiment = MEI(y_pred)
        elif method == MU:
            index_new_experiment = MU(y_std)
        elif method == MLI:
            index_new_experiment = MLI(y_pred, y_std)
        X_train, y_train, X_test, y_test, iteration, x_times_choosen = building_new_training_and_test_sets(X_train, y_train, X_test, y_test, index_new_experiment, iteration, x_times_choosen)    
    iterations_needed = iteration
    return (iterations_needed, X_train, y_train, x_times_choosen)

#two supportfunctions
def find(string, ch): 
    index = 0 
    while index < len(string): 
        if string[index] == ch: 
            return index 
        index = index + 1 
    return -1
def find_number(string, index):
    last_index = index
    number_string = ''
    while (string[last_index] == "0") or (string[last_index] == ".") or (string[last_index] == "1") or (string[last_index] == "2") or (string[last_index] == "3") or (string[last_index] == "4") or (string[last_index] == "5") or (string[last_index] == "6") or (string[last_index] == "7") or (string[last_index] == "8") or (string[last_index] == "9"):
        number_string = number_string + string[last_index]
        last_index = last_index + 1
    return number_string

def process_csv_datei(datapath):
    """create new columns for the elements in the compound"""
    csvdatei = pd.read_csv(datapath)
    Matrix = np.zeros((437,9))
    for i in range(437):
        string = csvdatei["Chemical formula"][i]
        if "Fe" in string:
            Matrix[i][0] = find_number(string, 5)
        if find(string, "C") > 0:
            Matrix[i][1] = find_number(string, find(string, "C")+4)
        if find(string, "n") > 0:
            Matrix[i][2] = find_number(string, find(string, "n")+4)
        if find(string, "u") > 0:
            Matrix[i][3] = find_number(string, find(string, "u")+4)
        if find(string, "i") > 0:
            Matrix[i][4] = find_number(string, find(string, "i")+4)
        if find(string, "P") > 0:
            Matrix[i][5] = find_number(string, find(string, "P")+4)
    #     if find(string, "S") > 0:
    #         Matrix[i][6] = find_number(string, find(string, "S")+4)
        if find(string, "o") > 0:
            Matrix[i][6] = find_number(string, find(string, "o")+4)
        if find(string, "N") > 0:
            Matrix[i][7] = find_number(string, find(string, "N")+5)
        if find(string, "r") > 0:
            Matrix[i][8] = find_number(string, find(string, "r")+4)
        #Matrix[i][6] = 1-Matrix[i][0]-Matrix[i][1]-Matrix[i][2]-Matrix[i][3]-Matrix[i][4]-Matrix[i][5]-Matrix[i][7]-Matrix[i][8]-Matrix[i][9]
    csvdatei = pd.concat([pd.DataFrame(Matrix), csvdatei], axis=1)
    #csvdatei = csvdatei.rename(columns={"0": "Fe", "1": "C", "2": "Mn", "3": "Cu", "4": "Si", "5": "P", "6": "S", "7": "Mo", "8": "Ni", "9": "Cr"})
    return csvdatei
    
```

```python
# doing FUELS framework with each method (MEI,MU,MLI) 30 times and 
# calculating the sample mean and uncertainty of the number of steps required to find the optimal candidat
# how many times out of these 30 times the method was conducted was a experiment choosen
# plotting a boxplot
# and a heatmap
total_iterations_MEI = 0
iterations_MEI = []
x_times_choosen_MEI = np.zeros((437, 1))
for i in range(30):
    csvdatei = process_csv_datei("citrination-export (1).csv") 
    X = csvdatei.drop(["Fatigue Strength","Chemical formula","Sample Number"],axis=1)
    y = csvdatei[["Fatigue Strength"]]
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=10, random_state=i)

    y = y.values
    X_train = X_train.values
    y_train = np.ravel(y_train)
    X_test = X_test.values
    y_test = np.ravel(y_test)
    iterations_needed, X_train, y_train, x_times_choosen = finding_best_candidate(y, y_train, y_test, X_train, X_test, MEI)
    x_times_choosen_MEI = x_times_choosen_MEI + x_times_choosen
    total_iterations_MEI = total_iterations_MEI + iterations_needed
    iterations_MEI.append(iterations_needed+10)
iterations_mean_MEI = total_iterations_MEI/30 +10
var_MEI = np.sum((np.array(iterations_MEI)-iterations_mean_MEI)**2)/(30-1)
std_MEI = np.sqrt(var_MEI)
uncertainty_MEI = std_MEI/np.sqrt(30)
print(iterations_mean_MEI)
print(uncertainty_MEI)
```

```python
total_iterations_MU = 0
iterations_MU = []
x_times_choosen_MU = np.zeros((437, 1))
for i in range(30):
    csvdatei = process_csv_datei("citrination-export (1).csv") 
    X = csvdatei.drop(["Fatigue Strength","Chemical formula","Sample Number"],axis=1)
    y = csvdatei[["Fatigue Strength"]]
    X_train, X_test, y_train, y_test  = train_test_split(X, y, train_size=10, random_state=i)

    y = y.values
    X_train = X_train.values
    y_train = np.ravel(y_train)
    X_test = X_test.values
    y_test = np.ravel(y_test)
    iterations_needed, X_train, y_train, x_times_choosen = finding_best_candidate(y, y_train, y_test, X_train, X_test, MU)
    x_times_choosen_MU = x_times_choosen_MU + x_times_choosen
    total_iterations_MU = total_iterations_MU + iterations_needed
    iterations_MU.append(iterations_needed+10)
iterations_mean_MU = total_iterations_MU/30 +10
var_MU = np.sum((np.array(iterations_MU)-iterations_mean_MU)**2)/(30-1)
std_MU = np.sqrt(var_MU)
uncertainty_MU = std_MU/np.sqrt(30)
print(iterations_mean_MU)
print(uncertainty_MU)
```

```python
total_iterations_MLI = 0
iterations_MLI = []
x_times_choosen_MLI = np.zeros((437, 1))
for i in range(30):
    csvdatei = process_csv_datei("citrination-export (1).csv") 
    X = csvdatei.drop(["Fatigue Strength","Chemical formula","Sample Number"],axis=1)
    y = csvdatei[["Fatigue Strength"]]
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=10, random_state=i)

    y = y.values
    X_train = X_train.values
    y_train = np.ravel(y_train)
    X_test = X_test.values
    y_test = np.ravel(y_test)
    iterations_needed, X_train, y_train, x_times_choosen = finding_best_candidate(y, y_train, y_test, X_train, X_test, MLI)
    x_times_choosen_MLI = x_times_choosen_MLI + x_times_choosen
    total_iterations_MLI = total_iterations_MLI + iterations_needed
    iterations_MLI.append(iterations_needed+10)
iterations_mean_MLI = total_iterations_MLI/30 +10
var_MLI = np.sum((np.array(iterations_MLI)-iterations_mean_MLI)**2)/(30-1)
std_MLI = np.sqrt(var_MLI)
uncertainty_MLI = std_MLI/np.sqrt(30)
print(iterations_mean_MLI)
print(uncertainty_MLI)
```

```python
# boxplot - how many iterations in average needed to find experiment optimal targetvalue (starting with 10 experiments)
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
iterations = [iterations_mean_MLI,iterations_mean_MEI,iterations_mean_MU]
uncertainties = [uncertainty_MLI,uncertainty_MEI,uncertainty_MU]
couleur = ['#3776ab', '#5bc48b', '#dec948']


fig, ax = plt.subplots()
plt.bar(x, iterations, color= couleur, yerr= uncertainties)
plt.xticks(x, ('MLI', 'MEI', 'MU'))
plt.show()
```

```python
#heatmap of the experiments
from sklearn.manifold import TSNE
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import Colormap

def prepare_heatmap(X,y,x_times_choosen):
    X_Method = X.values
    for h in reversed(range(len(x_times_choosen))):
        if x_times_choosen[h] == 0:
            X_Method = np.delete(X_Method, h, 0)
            y = np.delete(y, h, 0)
            x_times_choosen = np.delete(x_times_choosen, h, 0)
    return X_Method, y ,x_times_choosen

X_MLI, y_MLI ,x_times_choosen_MLI = prepare_heatmap(X,y,x_times_choosen_MLI)
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(X_MLI)
fig, ax = plt.subplots(figsize=(10,7))
sns.scatterplot(x=tsne_results[:,0], y=tsne_results[:,1], hue = x_times_choosen_MLI.reshape(len(x_times_choosen_MLI),),size=y_MLI.reshape(len(y_MLI),), ax=ax, sizes=(50, 300))
ax.set_title("Fatigue Strength and times of experiment choosen vs the two components of TSNE for MLI")
ax.legend(loc="upper left", bbox_to_anchor=(1,1))
plt.show()


X_MEI,y_MEI,x_times_choosen_MEI = prepare_heatmap(X,y,x_times_choosen_MEI)
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(X_MEI)
fig, ax = plt.subplots(figsize=(10,7))
sns.scatterplot(x=tsne_results[:,0], y=tsne_results[:,1], hue = x_times_choosen_MEI.reshape(len(x_times_choosen_MEI),),size=y_MEI.reshape(len(y_MEI),), ax=ax, sizes=(50, 300))
ax.set_title("Fatigue Strength and times of experiment choosen vs the two components of TSNE for MEI")
ax.legend(loc="upper left", bbox_to_anchor=(1,1))
plt.show()


X_MU,y_MU,x_times_choosen_MU = prepare_heatmap(X,y,x_times_choosen_MU)
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(X_MU)
fig, ax = plt.subplots(figsize=(10,7))
sns.scatterplot(x=tsne_results[:,0], y=tsne_results[:,1], hue = x_times_choosen_MU.reshape(len(x_times_choosen_MU),),size=y_MU.reshape(len(y_MU),), ax=ax, sizes=(50, 300))
ax.set_title("Fatigue Strength and times of experiment choosen vs the two components of TSNE for MU")
ax.legend(loc="upper left", bbox_to_anchor=(1,1))
plt.show()
```
